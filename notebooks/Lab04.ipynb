```python
# Create a Spark Session
spark = SparkSession.builder \
    .appName("MySQL") \
    .master("spark://localhost:7077") \
    .config("spark.jars", "/home/jovyan/work/mysql-connector-j-9.5.0.jar") \
    .getOrCreate()

# Thông tin kết nối MySQL
url = "jdbc:mysql://mysql:3306/sakila"
table = "film"  # Tên bảng cần đọc
user = "sakila_user"
password = "root123"

# Đọc dữ liệu từ bảng 'film'
film_df = spark.read.format("jdbc") \
    .option("url", url) \
    .option("dbtable", table) \
    .option("user", user) \
    .option("password", password) \
    .option("driver", "com.mysql.cj.jdbc.Driver") \
    .load()

# Hiển thị dữ liệu
film_df.show(10)
```
